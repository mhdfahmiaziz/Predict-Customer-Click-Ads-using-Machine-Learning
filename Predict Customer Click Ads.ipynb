# Data Cleansing

### Missing Values

```python
df.isnull().sum()
```

```python
# Imputation
df['Daily Time Spent on Site'] = df['Daily Time Spent on Site'].fillna(df['Daily Time Spent on Site'].median())
df['Area Income'] = df['Area Income'].fillna(df['Area Income'].mean())
df['Daily Internet Usage'] = df['Daily Internet Usage'].fillna(df['Daily Internet Usage'].mean())

df['Male'] = df['Male'].fillna(df['Male'].mode()[0])
```

### Inconsistent data

```python
df.info()
```

```python
# Rename columns Male
df.rename(columns = {'Male':'Gender'}, inplace = True)
```

```python
# Replace value Clicked on ad
df['Clicked on Ad'] = df['Clicked on Ad'].replace({'No' : 0, 'Yes' : 1})
```

```python
df.head()
```

```python
# Change tipe data Timestamp
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
```

```python
df['year'] = df['Timestamp'].dt.year
df['month'] = df['Timestamp'].dt.month
df['day'] = df['Timestamp'].dt.day
df['week'] = df['Timestamp'].dt.week
```

```python
df.info()
```

### Duplicated

```python
df.duplicated().any()
```

# Preprocessing

```python
df_encoding = df.copy()
```

### Feature Encoding

```python
cats2 = (df.select_dtypes('object').columns)
```

```python
df_encoding[cats2].describe()
```

```python
one_hot = ['Gender','province','category']

for cat in one_hot:
    onehots = pd.get_dummies(df_encoding[cat], prefix=cat)
    df_encoding = df_encoding.join(onehots)
    
df_encoding = df_encoding.drop(columns=['Gender','province','category'])

# check dataframe after encoding
df_encoding.info()
```

```python
df_encoding = df_encoding.drop(columns= ['city','Timestamp'])
```

```python
df_encoding.head()
```

# Split Train Test

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_encoding.drop('Clicked on Ad', axis=1)
y = df_encoding['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

# Modelling

```python
# Create function  model evaluation

def eval_classification(model,  y_train_pred, y_test_pred, y_train, y_test):
    print("Accuracy (Train Set): %.2f" % accuracy_score(y_train, y_train_pred))
    print("Accuracy (Test Set): %.2f" % accuracy_score(y_test, y_test_pred))
    print("Precision (Train Set): %.2f" % precision_score(y_train, y_train_pred))
    print("Precision (Test Set): %.2f" % precision_score(y_test, y_test_pred))
    print("Recall (Train Set): %.2f" % recall_score(y_train, y_train_pred))
    print("Recall (Test Set): %.2f" % recall_score(y_test, y_test_pred))
    print("F1-Score (Train Set): %.2f" % f1_score(y_train, y_train_pred))
    print("F1-Score (Test Set): %.2f" % f1_score(y_test, y_test_pred))
    
    fpr, tpr, thresholds = roc_curve(y_train, y_train_pred, pos_label=1) # pos_label: label yang kita anggap positive
    print("AUC (Train Set): %.2f" % auc(fpr, tpr))
    fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label=1) # pos_label: label yang kita anggap positive
    print("AUC (Test Set): %.2f" % auc(fpr, tpr))
```

# Business Recomendation

## Feature Based

Based on EDA and feature importance, we can conclude that:
1. The data we get has 2 user segments, namely the upper class and lower class users.
- The upper class has the criteria of frequently using the internet, frequently visiting a product's website, having a relatively young age and high income.
- The lower class has the opposite character.
2. Users with lower economic class, tend to be more easily attracted to product clicks on digital ads.
3. Users who are very heavy in using the internet are even more difficult to provide advertisements because they may be used to digital ads. 
4. Parents are a potential market for the digital market.

Business takeaway:
1. We can use a more unique way (soft selling) so that users don't see advertising as much.
2. Use mainstream content (simple but a topic of conversation) in order to hook users with the lower class segment.
